```python
import zipfile
import os

zip_path = '/content/archive.zip'
extract_path = '/content/extracted_data'

if os.path.exists(zip_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
```

```python
import yaml
from ultralytics import YOLO

config = {
    'path': '/content/extracted_data',
    'train': 'train/images',
    'val': 'valid/images',
    'test': 'test/images',
    'nc': 1,
    'names': ['license-plate']
}

with open('/content/extracted_data/data.yaml', 'w') as f:
    yaml.dump(config, f)

model_base = YOLO('yolo11n.pt')

model_base.train(
    data='/content/extracted_data/data.yaml',
    epochs=10,
    imgsz=640,
    batch=16,
    project='Lab1',
    name='baseline'
)
```

```python
model_improved = YOLO('yolo11s.pt')
model_improved.train(
    data='/content/extracted_data/data.yaml',
    epochs=20,
    imgsz=640,
    project='Lab1',
    name='improved'
)
```
